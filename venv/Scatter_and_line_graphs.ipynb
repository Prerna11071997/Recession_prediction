```python
import matplotlib.pyplot as plt
import pandas as pd
final = "final_dataset/final_part4.csv"
rec = pd.read_csv(final)
print(rec.head())
```

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(rec['GDP'], rec['Recession State'])
# set a title and labels
ax.set_title('Recession Dataset')
ax.set_xlabel('GDP')
ax.set_ylabel('Recession State')
```

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(rec['GDP'], rec['EXCH'])
# set a title and labels
ax.set_title('Recession Dataset')
ax.set_xlabel('GDP')
ax.set_ylabel('EXCH')
```

```python
# create a figure and axis
fig, ax = plt.subplots()

# scatter the sepal_length against the sepal_width
ax.scatter(rec['STOCK'], rec['GOLD'])
# set a title and labels
ax.set_title('Recession Dataset')
ax.set_xlabel('STOCK')
ax.set_ylabel('GOLD')
```

```python
# get columns to plot
columns = rec.columns.drop(['Unnamed: 0'])
# create x data
x_data = range(0, rec.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, rec[column], label=column)
# set title and legend
ax.set_title('Recession Dataset')
ax.legend()
```
